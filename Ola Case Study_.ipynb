Q(d). Determine if quarterly rating has increased (1 if yes, 0 if no).

```python
df["MMM-YY"]  = pd.to_datetime(df["MMM-YY"], format = "%d/%m/%y")
df = df.sort_values(by = ["Driver_ID", "MMM-YY"])

def Rating_increaing(rating):
    return (rating > rating.shift(1)).astype(int)
    

df["Rating_Increased"] = df.groupby("Driver_ID")["Quarterly Rating"].apply(Rating_increaing).reset_index(level = 0, drop = True)
```
